```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
categorical_cols = ['strain', 'host', 'province', 'country', 'status']
for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Count Plot of {col}')
    plt.show()
```

```python
# Analyze missing values visually
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# sns.countplot(x='brd_status', data=df, palette='coolwarm')
# sns.countplot(x='correct_column_name', data=df, palette='coolwarm')
df.rename(columns={'wrong_column_name': 'brd_status'}, inplace=True)


plt.title('Distribution of BRD Status')
plt.show()
```

```python
# Check if the column exists
if 'brd_status' in df.columns:
    # Count the unique values in brd_status
    print("BRD Status Distribution:")
    print(df['brd_status'].value_counts())
    
    # Plot the distribution of brd_status
    plt.figure(figsize=(10, 6))
    sns.countplot(x='brd_status', data=df, palette='coolwarm')
    plt.title('Distribution of BRD Status')
    plt.xlabel('BRD Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'brd_status' not found. Please check the column name.")
```

```python
continuous_features = ['year', 'isolate', 'clade']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(continuous_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(y=df[feature], color='skyblue')
    plt.title(f'{feature} Boxplot')
plt.tight_layout()
plt.show()
```

```python
continuous_features = ['year', 'isolate', 'clade']
z_scores = np.abs(zscore(df[continuous_features]))

outlier_indices = np.where(z_scores > 3)
print(f"\nNumber of outliers detected: {len(outlier_indices[0])}")
```
